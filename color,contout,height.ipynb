```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.DataFrame({"x":np.random.randn(10000),
                   "y":np.random.randn(10000),
                   "value":np.random.randn(10000)})
cmap="viridis"
alpha=1
plt.figure(figsize=(6,6))
plt.scatter(data["x"],data["y"],data["value"],cmap=cmap,alpha=alpha)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatterplot with colormap")
plt.colorbar(label="value")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.DataFrame(np.random.rand(50,50))
plt.figure(figsize=(8,6))
plt.contour(data.values)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
data=pd.DataFrame(np.random.rand(3,3))
x=np.arange(data.shape[1])
y=np.arange(data.shape[1])
x,y=np.meshgrid(x,y)
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(x,y,data.values,cmap='viridis')
plt.show()
```
