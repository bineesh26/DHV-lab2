```python
import folium

location = [13.169246990000547, 77.5348567563776]

mymap = folium.Map(location=location,
                   width=950,
                   height=550,
                   zoom_start=12,
                   tiles='openstreetmap')

folium.Marker(
    location=location,
    popup='Presidency University, Bangalore',
    icon=folium.Icon(color='blue')
).add_to(mymap)

mymap.save('presidency_university_map.html')
mymap
```
